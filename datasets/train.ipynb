```python
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import yaml
import os
import cv2
import random

best_model = YOLO('runs/detect/train2/weights/best.pt')

try:
    results_csv_path = 'runs/detect/train2/results.csv'
    metrics_df = pd.read_csv(results_csv_path)
    plt.figure(figsize=(16, 12))
    
    # Training losses
    plt.subplot(2, 2, 1)
    plt.plot(metrics_df['train/box_loss'], label='Box Loss')
    plt.plot(metrics_df['train/cls_loss'], label='Class Loss')
    plt.plot(metrics_df['train/dfl_loss'], label='DFL Loss')
    plt.title('Training Losses')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    # Validation metrics
    plt.subplot(2, 2, 2)
    plt.plot(metrics_df['metrics/precision(B)'], label='Precision')
    plt.plot(metrics_df['metrics/recall(B)'], label='Recall')
    plt.plot(metrics_df['metrics/mAP50(B)'], label='mAP50')
    plt.plot(metrics_df['metrics/mAP50-95(B)'], label='mAP50-95')
    plt.title('Validation Metrics')
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    
    # Learning rate
    plt.subplot(2, 2, 3)
    plt.plot(metrics_df['lr/pg0'], label='Learning Rate')
    plt.title('Learning Rate')
    plt.xlabel('Epoch')
    plt.ylabel('LR')
    plt.grid(True)
    
    # Validation losses
    plt.subplot(2, 2, 4)
    plt.plot(metrics_df['val/box_loss'], label='Val Box Loss')
    plt.plot(metrics_df['val/cls_loss'], label='Val Class Loss')
    plt.plot(metrics_df['val/dfl_loss'], label='Val DFL Loss')
    plt.title('Validation Losses')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('training_metrics.png')
    plt.show()
    
except Exception as e:
    print(f"Error plotting training metrics: {e}")

print("\nRunning validation...")
val_results = best_model.val(data='data.yaml')

try:
    with open('data.yaml', 'r') as f:
        data_config = yaml.safe_load(f)
        class_names = data_config['names']
        print(f"Classes: {class_names}")
except Exception as e:
    print(f"Error loading class names: {e}")
    class_names = [f"Class {i}" for i in range(9)]  

print("\nGenerating confusion matrix...")
try:
    conf_matrix = val_results.confusion_matrix.matrix
    
    row_sums = conf_matrix.sum(axis=1, keepdims=True)
    row_sums = np.where(row_sums == 0, 1, row_sums)  
    conf_matrix_norm = conf_matrix / row_sums
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(conf_matrix_norm, annot=True, fmt='.2f', cmap='Blues',
               xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Normalized Confusion Matrix')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()
    
except Exception as e:
    print(f"Error generating confusion matrix: {e}")

print("\nGenerating sample predictions...")
try:
    valid_paths = ['valid/images', 'data/valid/images', 'dataset/valid/images', 
                  '../valid/images', '../data/valid/images']
    
    valid_dir = None
    for path in valid_paths:
        if os.path.exists(path):
            valid_dir = path
            break
            
    if valid_dir:
        print(f"Found validation images in: {valid_dir}")
        valid_images = [os.path.join(valid_dir, img) for img in os.listdir(valid_dir) 
                       if img.endswith(('.jpg', '.jpeg', '.png'))]
        
        num_samples = min(3, len(valid_images))
        if num_samples > 0:
            sample_images = random.sample(valid_images, num_samples)
            
            plt.figure(figsize=(15, 5*num_samples))
            for i, img_path in enumerate(sample_images):
                results = best_model(img_path)
                
                result_image = results[0].plot()
                result_image = cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB)

                plt.subplot(num_samples, 1, i+1)
                plt.imshow(result_image)
                plt.title(f'Sample Prediction: {os.path.basename(img_path)}')
                plt.axis('off')
            
            plt.tight_layout()
            plt.savefig('sample_predictions.png')
            plt.show()
    else:
        print("Could not find validation images directory.")
        
except Exception as e:
    print(f"Error generating sample predictions: {e}")

print("\nModel Performance Summary:")
print("--------------------------")
print(f"mAP50: {val_results.box.map50:.4f}")
print(f"mAP50-95: {val_results.box.map:.4f}")
print(f"Precision: {val_results.box.mp:.4f}")
print(f"Recall: {val_results.box.mr:.4f}")

metrics_summary = pd.DataFrame({
    'Metric': ['mAP50', 'mAP50-95', 'Precision', 'Recall'],
    'Value': [
        val_results.box.map50,
        val_results.box.map,
        val_results.box.mp,
        val_results.box.mr
    ]
})
metrics_summary.to_csv('model_metrics_summary.csv', index=False)
print("\nMetrics summary saved to model_metrics_summary.csv")
print("Visualizations saved as PNG files")
```
